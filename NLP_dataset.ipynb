# Data Trimming

Due to the Huge Size of the Dataset, I have chosen to focus specifically on Laptop Product Reviews. In this cell, I read in the Product List and filter out products that are not laptop or tablet. The price is used as a filter to filter out laptop accessories like laptop sleeves that are also included in the original dataset. The rating number is used to reduce the final dataset to the popular products.

```python
import json

counter = 0
file = 'C:/Users/kwh/Documents/NLP/meta_Electronics.jsonl'

products = []

with open(file, 'r') as fp:
    for line_num, line in enumerate(fp, 1):
        try:
            item = json.loads(line.strip())
            if item["main_category"] == "Computers":
                new_item = {'title': item['title'],
                            'average_rating': item['average_rating'],
                            'rating_number': item['rating_number'],
                            'features': item['features'],
                            'description': item['description'],
                            'price': item['price'],
                            'parent_asin': item['parent_asin']}
                if ('laptop' in item['title'].lower() or 'tablet' in item['title'].lower()) and item['rating_number'] > 5000 and item['price'] and item['price'] > 200:
                    counter += 1
                    products.append(new_item)
        except json.JSONDecodeError as e:
            print(f"Skipping malformed JSON line {line_num}: {line.strip()} - Error: {e}")
print(counter)
```

## Write the filtered Dataset to a JSON

```python
import json

output_file = 'meta_electroics.json'

with open(output_file, 'w') as fp:
    json.dump(products, fp, indent=4)

print(f"Successfully wrote {len(products)} products to {output_file}")
```

## Load the Product Json and select Reviews for these Products

```python
import json
file = 'C:/Users/kwh/Documents/NLP/meta_electroics.json'

with open(file, 'r') as f:
    data = json.load(f)

products_asin = []
for product in data:
    products_asin.append(product['parent_asin'])
```

```python
file = 'C:/Users/kwh/Documents/NLP/Electronics.jsonl'

reviews = []
with open(file, 'r') as f:
  for line in f:
    review = json.loads(line.strip())
    if review['parent_asin'] in products_asin:
        reviews.append(review)

print(len(reviews))
```

```python
import json

output_file = 'electronics.json'

with open(output_file, 'w') as f:
    json.dump(reviews, f, indent=4)

print(f"Successfully wrote {len(reviews)} reviews to {output_file}")
```
